## Data cleaning

```python
# assigning tweet column to'tweets' variable
tweets= df.tweet
tweets
```

```python
# change tweets to lower case
to_lower=lambda x: x.lower()
tweets_a = tweets.apply(to_lower)
tweets_a
```

```python
# remove tweet tags, 'rt' and newlines
remove_stopwords= lambda x: ' '.join([word for word in x.split() if not word.startswith(('@', 'rt', '\n'))]) #.join is from package string
tweets_b = tweets_a.apply(remove_stopwords)
tweets_b
```

```python
# remove urls
remove_url= lambda x:re.sub(r'http\S+','',str(x))
tweets_c = tweets_b.apply(remove_url)
tweets_c
```

```python
# remove punctuations
remove_puncs= lambda x:x.translate(str.maketrans('','',string.punctuation))
tweets_c = tweets_b.apply(remove_puncs)
tweets_c
```

```python
# remove common words (stop words)
more_words=["i", "me", "my", "myself", "we", "our", "ours", "ourselves", "you", "your", "yours", "yourself", "yourselves", "he", "him", "his", "himself", "she", "her", "hers", "herself", "it", "its", "itself", "they", "them", "their", "theirs", "themselves", "what", "which", "who", "whom", "this", "that", "these", "those", "am", "is", "are", "was", "were", "be", "been", "being", "have", "has", "had", "having", "do", "does", "did", "doing", "a", "an", "the", "and", "but", "if", "or", "because", "as", "until", "while", "of", "at", "by", "for", "with", "about", "against", "between", "into", "through", "during", "before", "after", "above", "below", "to", "from", "up", "down", "in", "out", "on", "off", "over", "under", "again", "further", "then", "once", "here", "there", "when", "where", "why", "how", "all", "any", "both", "each", "few", "more", "most", "other", "some", "such", "no", "nor", "not", "only", "own", "same", "so", "than", "too", "very", "s", "t", "can", "will", "just", "don", "should", "now"]
stop_words=set(stopwords.words('english')) #nltk package
stop_words.update(more_words)

remove_stopwords=lambda x: ' '.join([word for word in x.split() if word not in stop_words]) #.join is from package string
tweets_d = tweets_c.apply(remove_stopwords)
tweets_d
```

```python
# remove emoticons and emojis
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
tweets_e = tweets_d.apply(lambda x: remove_emoji(x))
tweets_e
```

```python
# check if columns contain non ascii (english) letters
(~tweets_e.map(str.isascii)).any()
```

```python
# show columns that contain non ascii (english) letters
tweets_e[~tweets_e.map(str.isascii)]
```

```python
# Remove non ascii letters from tweets
tweets_f = tweets_e.str.encode('ascii', 'ignore').str.decode('ascii')
(~tweets_f.map(str.isascii)).any()
```

```python
# add clean data to dataframe
df['tweet'] = tweets_f
df.head(3)
```
